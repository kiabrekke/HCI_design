```python
pubs_df = pd.read_csv('open_pubs.csv', dtype={'fas_id' : int, 'name': str, 'address' : str, 'postcode': str, 'easting' : float, 'northing' : float, 'latitude': str, 'longitude': str, 'local_authority': str})
pubs_df
```

```python
#number of pubs in the whole set
print('Pubs in England : '+ str(pubs_df.shape[0]))

#sorting out only pubs in Leeds + number of pubs in Leeds
leeds_pubs_df = pubs_df.loc[pubs_df['local_authority'] == 'Leeds']
print('Pubs in Leeds : ' + str(leeds_pubs_df.shape[0]))

leeds_pubs_df.info()
```

```python
leeds_before_drop = leeds_pubs_df
leeds_before_drop.head()
list(leeds_before_drop)
```

```python
#drop columns
leeds_necc = leeds_pubs_df[['name', 'easting', 'northing']]
#newdf = df[df.columns[2:4]]
leeds_necc.head()
```

```python
#create TSV file
leeds_before_drop.to_csv('data.tsv', header = False, index = False,  sep = '\t')
leeds_pubs_df.to_csv('data_pubs.csv', header = False, index = True)
```

```python
# bokeh basics
from bokeh.plotting import figure
from bokeh.io import show, output_notebook
from bokeh.models import ColumnDataSource

# Create a blank figure with labels
p = figure(plot_width = 600, plot_height = 600, 
           title = 'Example Glyphs',
           x_axis_label = 'X', y_axis_label = 'Y')

# Plot data
circles_x = leeds_necc['easting'].tolist()
circles_y = leeds_necc['northing'].tolist()

# Add circle glyph
p.circle(circles_x, circles_y, size = 10, color = 'red')


# Set to output the plot in the notebook
output_notebook()
# Show the plot
show(p)
```
